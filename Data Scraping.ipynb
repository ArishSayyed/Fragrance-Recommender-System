```python
# create an Empty DataFrame 
# object With column names only 
main_df = pd.DataFrame(columns = ['Fragrance', 'By', 'Image', 'Description']) 
main_df.head()
```

```python
data = []
for i in range(0,len(soup.find('div',{'class':'search__results'}).findAll('div',{'itemtype':'http://schema.org/SomeProducts'}))):
    temp_dict = {}
    fragrance_name =''
    by =''
    img_url = ''
    description =''
    #Fragrance name
    fragrance_name = soup.find("div",{"class":"search__results"}).findAll('div',{"itemtype":"http://schema.org/SomeProducts"})[i].h2.text
    #by
    by = soup.find("div",{"class":"search__results"}).findAll('div',{"itemtype":"http://schema.org/SomeProducts"})[i].p.a.text
    #img_url
    img_url = soup.find("div",{"class":"search__results"}).findAll('div',{"itemtype":"http://schema.org/SomeProducts"})[i].find('img')['data-src']
    
    #fragrance_url
    fragrance_url = soup.find("div",{"class":"search__results"}).findAll('div',{"itemtype":"http://schema.org/SomeProducts"})[i].find("a")['href']
    fragrance_url = url[0:23] + fragrance_url
    """visiting page of every fragrance one by one to get description"""
    driver = webdriver.Chrome('chromedriver')
    driver.get(fragrance_url)
    fragrance_page_source = driver.page_source
    time.sleep(2)
    driver.quit()
    #giving it to bs
    fragrance_soup = BeautifulSoup(fragrance_page_source, 'html.parser')
    """Scraping description"""
    desc1 = fragrance_soup.find('span', {'itemprop':"Description"}).text
    try:
        desc2 = fragrance_soup.find('span', {'id':'product-read-more-text-mq1'}).text
    except:
        desc2 = ''
    try:
        desc3 = fragrance_soup.find('span', {'id':'product-read-more-text'}).text
    except:
        desc3 = ''
    finally:
        description = desc1 + desc2 + desc3
        description = re.sub('\n', '', description)
    
    #Dictionary
    temp_dict = {'Fragrance':fragrance_name, 'By': by, 'Image':img_url, 'Description':description}
    data.append(temp_dict)
    
    
data_df = pd.DataFrame.from_dict(data)
```

```python
main_df = pd.concat([main_df, data_df], ignore_index=True)
```
